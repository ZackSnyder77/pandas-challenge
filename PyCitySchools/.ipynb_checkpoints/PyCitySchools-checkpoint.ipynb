```python
import pandas as pd
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
#Creating the needed Series to obtain the answer and values requested
total_schools = len(school_data)
total_students = len(school_data_complete)
budget = school_data["budget"].sum()
avg_math = school_data_complete["math_score"].mean()
avg_read = school_data_complete["reading_score"].mean()
pass_rate = (avg_math + avg_read)/2

math_condition = school_data_complete["math_score"]>=70
math_pass_students = school_data_complete[math_condition]
math_pass_rate = (len(math_pass_students)/total_students)*100


read_condition = school_data_complete["reading_score"]>=70
read_pass_students = school_data_complete[read_condition]
read_pass_rate = (len(read_pass_students)/total_students)*100
```

```python
# Creating a dataframe to hold the above values and formating for a cleaner output

summary_df = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": total_students,
    "Total Budget": budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    "% Passing Math": math_pass_rate,
    "% Passing Reading": read_pass_rate,
    "% Overall Passing Rate": pass_rate
    
})

summary_df["Total Students"] = summary_df["Total Students"].astype(float).map("{:,.0f}".format)
summary_df["Total Budget"] = summary_df["Total Budget"].astype(float).map("${:,.2f}".format)
summary_df
```
